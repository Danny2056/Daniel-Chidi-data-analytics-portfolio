```python
def calculate_risk_score(support_tickets, days_since_login, payment_status):
    score = 0

    if support_tickets >= 5:
        score += 2
    elif support_tickets >= 3:      # 3 to 4
        score += 1

    if days_since_login > 60:
        score += 2
    elif days_since_login >= 30:    # 30 to 60
        score += 1

    if payment_status == "Overdue":
        score += 2

    return score
```

```python
# Test cases
print(calculate_risk_score(6, 75, "Overdue"))   # 2+2+2 = 6
print(calculate_risk_score(4, 35, "Paid"))       # 1+1+0 = 2
print(calculate_risk_score(1, 8,  "Paid"))       # 0+0+0 = 0
print(calculate_risk_score(5, 60, "Overdue"))    # 2+1+2 = 5  (60 is "30 to 60", not >60)
```

```python
def classify_churn_risk(risk_score):
    if risk_score >= 5:
        return "High Risk"
    elif risk_score >= 3:           # 3 to 4
        return "Medium Risk"
    else:                           # 0 to 2
        return "Low Risk"
```

```python
def clean_customer_name(name):
    return name.strip().title()

def calculate_customer_value(monthly_fee, months_active):
    return monthly_fee * months_active

def classify_customer_value(customer_value):
    if customer_value >= 2000:
        return "Premium"
    elif customer_value >= 750:
        return "Standard"
    else:
        return "Basic"

def calculate_risk_score(support_tickets, days_since_login, payment_status):
    score = 0
    if support_tickets >= 5:
        score += 2
    elif support_tickets >= 3:
        score += 1
    if days_since_login > 60:
        score += 2
    elif days_since_login >= 30:
        score += 1
    if payment_status == "Overdue":
        score += 2
    return score

def classify_churn_risk(risk_score):
    if risk_score >= 5:
        return "High Risk"
    elif risk_score >= 3:
        return "Medium Risk"
    else:
        return "Low Risk"

print("All 5 helper functions defined.")
```

```python
def prepare_customer_retention_data(df):
    df = df.copy()

    df["customer_name"] = df["customer_name"].apply(clean_customer_name)

    df["customer_value"] = df.apply(
        lambda r: calculate_customer_value(r["monthly_fee"], r["months_active"]), axis=1)

    df["value_segment"] = df["customer_value"].apply(classify_customer_value)

    df["risk_score"] = df.apply(
        lambda r: calculate_risk_score(
            r["support_tickets"], r["days_since_login"], r["payment_status"]), axis=1)

    df["churn_risk"] = df["risk_score"].apply(classify_churn_risk)

    return df

prepared_data = prepare_customer_retention_data(customer_data)
prepared_data
```

```python
def get_high_risk_customers(df):
    columns = ["customer_id", "customer_name", "customer_value",
               "support_tickets", "days_since_login", "payment_status", "churn_risk"]
    return df[df["churn_risk"] == "High Risk"][columns]

get_high_risk_customers(prepared_data)
```

```python
def summarize_risk_by_segment(df):
    return (df.groupby(["value_segment", "churn_risk"])
              .size()
              .reset_index(name="customer_count"))

summarize_risk_by_segment(prepared_data)
```

```python
def create_follow_up_message(row):
    return (f"Contact {row['customer_name']}: {row['churn_risk']}. "
            f"Review account value of ${row['customer_value']}.")

# Apply to every customer
prepared_data.apply(create_follow_up_message, axis=1).tolist()
```

```python
print("HIGH RISK CUSTOMERS")
print(get_high_risk_customers(prepared_data)[["customer_name", "customer_value", "churn_risk"]])

print("\nRISK BY SEGMENT")
print(summarize_risk_by_segment(prepared_data))
```
